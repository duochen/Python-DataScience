```python
# Import packages
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Create plot values
x = np.linspace(0, 10)   # x-axis/time-scale
x # show x
```

```python
# Create corresponding cosine values from x
y = np.cos(x)
y # show y values
```

```python
# Create functional plot
plt.plot(x, y)
```

```python
# Create corresponding sine values from x
z = np.sin(x)
z # show z values
```

```python
# Plot y and z on x-axis
plt.plot(x,y,x,z);plt.show()
```

```python
# label the axes and give the graph title
plt.plot(x,y,x,z)
plt.xlabel('Time axis')  # label x-axis
plt.ylabel('Magnitude')  # label y-axis
plt.title('Since and Cosine waves')  # graph title
plt.show()    # print the graph
```

```python
# Add legend
plt.plot(x,y,x,z)
plt.xlabel('Time axis')
plt.ylabel('Magnitude')
plt.title('Sine and Cosine waves')
plt.legend(['y','z'],loc=0)  # loc=0 means best location
plt.show()
```

```python
# Use subplotting
plt.subplot(1,2,1)  # plot 1
plt.plot(x,y)
plt.title('Cosine plot')

plt.subplot(1,2,2)  # plot 2
plt.plot(x,z)
plt.title('Sine plot')

plt.tight_layout()  # avoid overallping plots
plt.show()
```

```python
# Change line and marker styles
plt.subplot(1,2,1)  # plot 1
plt.plot(x,y,'r-x') # red plot with -x marker
plt.title('Cosine plot')

plt.subplot(1,2,2)  # plot 2
plt.plot(x,y,'g-o') # green plot with -o marker
plt.title('Sine plot')

plt.tight_layout()   # avoiding overallping plots
plt.show()
```
