```python
# Create a DataFrame with Null values
df_null = pd.DataFrame({"name": ['Superman', 'Batman', 'Spiderman'],
                       "toy": [np.nan, 'Batmobile', 'Spiderman toy'],
                       "born": [pd.NaT, pd.Timestamp("1956-06-26"), pd.NaT]})
df_null
```

```python
# Check for null values, returns boolean array
df_null.isnull()
```

```python
# Opposite of pd.isnull()
df_null.notnull()
```

```python
# Drop all rows that contain null values
df_null.dropna()
```

```python
# Drop all columns that contain null values
df_null.dropna(axis=1)
```

```python
# Keep only the rows with at least 2 non-NA values
df_null.dropna(thresh=2)
```

```python
# Repalce all null values with x
df_null.fillna('x')
```

```python
# Create a DataFrame with missing numeric numbers
df_null_n = pd.DataFrame([[np.nan, 2, np.nan, 0], 
                          [3,4, np.nan, 1],
                          [np.nan, np.nan,np.nan,5],
                          [np.nan, 3, np.nan,4]],
                         columns=list('ABCD'))
df_null_n
```

```python
# Replace all NaN elements with 0s
df_null_n.fillna(0)
```

```python
# Replace all null values with the mean
df_null_n.fillna(df_null_n.mean())
```
