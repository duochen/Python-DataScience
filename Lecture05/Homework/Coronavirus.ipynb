```python
# Import the requried libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load confirmed cases
url_confirmed = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv'
confirmed = pd.read_csv(url_confirmed)
# Display the data
confirmed.head()
```

```python
# Load recovered cases
url_recovered = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv'
recovered = pd.read_csv(url_recovered)
# Display the data
recovered.head()
```

```python
# Load the death cases
url_death = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv'
deaths = pd.read_csv(url_death)
# Display the data
deaths.head()
```

```python
# Find out how many missing values in confirmed DataFrame
confirmed.isnull().sum()[confirmed.isnull().sum() > 0]
```

```python
# Find out how many missing values in recovered DataFrame
recovered.isnull().sum()[recovered.isnull().sum() > 0]
```

```python
# Find out how many missing values in deaths DataFrame
deaths.isnull().sum()[deaths.isnull().sum() > 0]
```

```python
# Replace the missings with 'unknown'
confirmed = confirmed.fillna('unknow')
recovered = recovered.fillna('unknow')
deaths = deaths.fillna('unknow')
```

```python
# Define the last update
last_update = '2/28/20'
```

```python
#  taking total confirmed, recovered and deaths from last_update and joing
china_cases = confirmed[['Province/State',last_update]][confirmed['Country/Region']=='Mainland China']
china_cases
```

```python
china_cases['recovered'] = recovered[last_update][recovered['Country/Region']=='Mainland China']
china_cases
```

```python
china_cases['deaths'] = deaths[last_update][deaths['Country/Region']=='Mainland China']
china_cases
```

```python
# setting "Province/State" as index
china_cases = china_cases.set_index('Province/State')
china_cases
```

```python
# renaming columns
china_cases = china_cases.rename(columns = {last_update:'confirmed','recovered':'recovered','deaths':'deaths'})
china_cases
```

```python
# creating the plot
china_cases.sort_values(by='confirmed',ascending=True).plot(kind='barh'
                                                            , figsize=(20,30)
                                                            , color = ['#4b8bbe','lime','red']
                                                            , width=1
                                                            , rot=2)

china_cases
```

```python
# defyning legend and titles parameters
plt.title('Total cases by Province/State in Mainland China', size=40)
plt.ylabel('Province/State',size=30)
plt.yticks(size=20)
plt.xticks(size=20)
plt.legend(bbox_to_anchor=(0.95,0.95) # setting coordinates for the caption box
           , frameon = True
           , fontsize = 20
           , ncol = 2 
           , fancybox = True
           , framealpha = 0.95
           , shadow = True
           , borderpad = 1);
plt.show()
```

```python
# Display Hubei cases distribution
# taking cases numbers
Hubei = china_cases[china_cases.index=="Hubei"]
Hubei = Hubei.iloc[0]
Hubei
```

```python
# difyning plot size
plt.figure(figsize=(15,15))

# here i use .value_counts() to count the frequency that each category occurs of dataset
Hubei.plot(kind='pie'
           , colors=['#4b8bbe','lime','red']
           , autopct='%1.1f%%' # adding percentagens
           , shadow=True
           , startangle=140)

# defyning titles and legend parameters
plt.title('Hubei Cases Distribution',size=30)
plt.legend(loc = "upper right"
           , frameon = True
           , fontsize = 15
           , ncol = 2 
           , fancybox = True
           , framealpha = 0.95
           , shadow = True
           , borderpad = 1);
plt.show()
```
