```python
df_full = df.dropna(subset=['AgeTrafficked'])
df_full.shape
```

```python
df_full.head(10)
```

```python
series = df_full['AgeTrafficked']
type(series)
```

```python
def convert_to_days(s):
    for item in s:
        last_ch = item[-1]
        if (last_ch) == 'd':
            item = item[:-1]
            num_item = round(float(item))
            item = num_item
            # print(f"days: {num_item}")
        if (last_ch) == 'm':
            item = item[:-1]
            num_item = round(float(item) * 30)
            item = num_item
            # print(f"days: {num_item}")
        if (last_ch) == 'y':
            item = item[:-1]
            num_item = round(float(item) * 365)
            item = num_item
            # print(f"days: {num_item}")
            
    return s
```

```python
new_series = convert_to_days(series)
df_full['NewAgeTrafficked']=new_series
df_full
```
