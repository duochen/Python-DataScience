```python
df_full = df_full.drop( df_full[ df_full['AgeTrafficked'] == '未出生' ].index , inplace=True)
```

```python
series = df_full['AgeTrafficked']
type(series)
```

```python
def convert_to_days(s):
    for item in s:
        try:
            if item == '7y-8':
                num_item = 7.5 * 365
                item = num_item
                print("7y-8", item)
            elif item.isnumeric():
                item = round(float(item))
            else:
                last_ch = item[-1]
                if (last_ch) == 'd':
                    item = item[:-1]
                    num_item = round(float(item))
                    item = num_item
                    # print(f"days: {num_item}")
                elif (last_ch) == 'm':
                    item = item[:-1]
                    num_item = round(float(item) * 30)
                    item = num_item
                    # print(f"days: {num_item}")
                elif last_ch.lower() == 'y':
                    item = item[:-1]
                    num_item = round(float(item) * 365)
                    item = num_item
                    # print(f"days: {num_item}")
                else:
                    print("Inside else: ", item)
        except ValueError:
            print("Got error: ", item)
            
    return s
```

```python
new_series = convert_to_days(series)
new_series
```
