```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/duochen/Python-DataScience/master/OtherDataSource/byprovince_byyear_Child_trafficked_by_year_by_region.csv'
df = pd.read_csv(url)
df.head(10)
```

```python
df_new = df.drop(df.index[0])
df_new
```

```python
df_index = df_new.set_index(df.columns[0])
df_index
```

```python
df_index.plot.bar(stacked=True)
```

```python
df_index.plot.area()
```

```python
for col_name in df_index.columns:
    print(col_name)
```

```python
df_final = pd.DataFrame(columns=['Region', 'Year', 'Trafficked'])
for index, row in df_index.iterrows():
    region = index
    for col in df_index.columns:
        year = col
        trafficked = row[col]
        a_row = {"Region": region, "Year": year, "Trafficked": trafficked}
        df_final = df_final.append(a_row, ignore_index=True)
                           
df_final
```

```python
url_grp = 'https://raw.githubusercontent.com/duochen/Python-DataScience/master/OtherDataSource/grp.csv'
df_grp = pd.read_csv(url_grp)
df_grp = df_grp.sort_values('Region')
df_grp
df_grp = df_grp.set_index('Region')
df_grp
```

```python
df_final
```

```python
df_region_grp = pd.DataFrame(columns=["Region", "Year", "GRP"])
for index, row in df_grp.iterrows():
    Region = index
    #print(f"Region: {Region}")
    for col in df_grp.columns:
        Year = col
        GRP = row[col]
        #print(f"Region: {Region}, Year: {Year}, GRP: {GRP}")
        a_row = {"Region": Region, "Year": Year, "GRP": GRP}
        df_region_grp = df_region_grp.append(a_row, ignore_index=True)
                 
df_region_grp
```

```python
df_region_grp.drop(["Region","Year"], axis=1, inplace=True)
```

```python
result = pd.concat([df_final, df_region_grp], axis=1, sort=False)
result
```
