```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/duochen/Python-DataScience/master/OtherDataSource/byprovince_byyear_Child_trafficked_by_year_by_region.csv'
df = pd.read_csv(url)
df.head(10)
```

```python
df_new = df.drop(df.index[0])
df_new
```

```python
df_index = df_new.set_index(df.columns[0])
df_index
```

```python
df_index.plot.bar(stacked=True)
```

```python
df_index.plot.area()
```

```python
for col_name in df_index.columns:
    print(col_name)
```

```python
df_final = pd.DataFrame(columns=['Region', 'Year', 'Trafficked'])
for index, row in df_index.iterrows():
    region = index
    for col in df_index.columns:
        year = col
        trafficked = row[col]
        a_row = {"Region": region, "Year": year, "Trafficked": trafficked}
        df_final = df_final.append(a_row, ignore_index=True)
                           
df_final
```
