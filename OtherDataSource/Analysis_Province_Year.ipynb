```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/duochen/Python-DataScience/master/OtherDataSource/byprovince_byyear_Child_trafficked_by_year_by_region.csv'
df = pd.read_csv(url)
df.head(10)
```

```python
df_new = df.drop(df.index[0])
df_new
```

```python
df_index = df_new.set_index(df.columns[0])
df_index
```

```python
df_index.plot.bar(stacked=True)
```

```python
df_index.plot.area()
```
